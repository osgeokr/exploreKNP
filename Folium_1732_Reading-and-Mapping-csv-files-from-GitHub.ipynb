```python
import folium
import pandas as pd
```

```python
url = (
    "https://raw.githubusercontent.com/osgeokr/exploreKNP/main/data/"
)
df = pd.read_csv(f'{url}/explore_list.csv', encoding='utf-8')
df.head()
```

```python
# 타일 유형
m = folium.Map(location=[36.83630855650386, 128.09758716104764], zoom_start=16, tiles="CartoDB positron")

# 마커
for index, row in df.iterrows():
    folium.Marker(
        location=[row['LAT'], row['LON']],
        popup=row['NAME'],
        icon=folium.Icon(color="purple", icon="info-sign"),
    ).add_to(m)
    
m
```

```python
# 사용자 위치 확인
from folium import plugins
plugins.LocateControl().add_to(m)
```
