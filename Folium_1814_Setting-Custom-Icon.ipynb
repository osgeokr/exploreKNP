```python
import folium
from folium import plugins
from folium.features import CustomIcon
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/osgeokr/exploreKNP/main/data/{}".format
df = pd.read_csv(url("explore_list.csv"), encoding='utf-8')
df.head()
```

```python
# Explorer(탐험가코스)
df_explorer = df[df['CAT_EN']=='Explorer']
df_explorer.head()
```

```python
# 타일 유형
m = folium.Map(location=[36.83630855650386, 128.09758716104764], zoom_start=16, tiles="CartoDB positron")

# 마커
for index, row in df_explorer.iterrows():
    folium.Marker(
        location=[row['LAT'], row['LON']],
        popup=row['NAME'],
        icon=folium.Icon(color="purple", icon="info-sign")
    ).add_to(m)
    
# 사용자 위치 확인
plugins.LocateControl().add_to(m)
    
m
```

```python
# 타일 유형
m = folium.Map(location=[36.83630855650386, 128.09758716104764], zoom_start=16, tiles="CartoDB positron")

icon_image = url("Explorer.png")

#Custom Icon (사용자 아이콘)
icon = CustomIcon(
    icon_image,
    icon_size=(200, 200)
)

marker = folium.Marker(
    location=[36.83630855650386, 128.09758716104764],
    icon=icon
)

m.add_child(marker)
m
```

```python
# 타일 유형
m = folium.Map(location=[36.83630855650386, 128.09758716104764], zoom_start=16, tiles="CartoDB positron")

icon_image = url("Explorer.png")

# 사용자 아이콘
for index, row in df_explorer.iterrows():
    icon = CustomIcon(icon_image, icon_size=(50, 50))
    marker = folium.Marker(location=[row['LAT'], row['LON']], icon=icon)
    m.add_child(marker)

# 사용자 위치 확인
plugins.LocateControl().add_to(m)

m
```

```python
# 파일로 저장
m.save("index.html")
```
