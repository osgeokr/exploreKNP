```python
import folium
from folium import plugins
from folium.features import CustomIcon
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/osgeokr/exploreKNP/main/data/{}".format
df = pd.read_csv(url("explore_list.csv"), encoding='utf-8')
df.head()
```

```python
print(f"고유값: {df['CAT_EN'].unique()}")
```

```python
df_explorer = df[df['CAT_EN']=='Explorer'] # Explorer(탐험가코스)
df_challenger = df[df['CAT_EN']=='Challenger'] # Challenger(도전자코스)
df_children = df[df['CAT_EN']=='Children'] # Children(어린이코스)
```

```python
# 타일 유형
m = folium.Map(location=[df.LAT.mean(), df.LON.mean()], zoom_start=16, tiles="CartoDB positron")

icon_image = url("Explorer.png")

# 사용자 아이콘
for index, row in df_explorer.iterrows():
    icon = CustomIcon(icon_image, icon_size=(50, 50))
    marker = folium.Marker(location=[row['LAT'], row['LON']], icon=icon)
    m.add_child(marker)

# 사용자 위치 확인
plugins.LocateControl().add_to(m)

m
```

```python
# 타일 유형
m = folium.Map(location=[df.LAT.mean(), df.LON.mean()], zoom_start=16, tiles="CartoDB positron")

# Explorer(탐험가코스)
f_explorer = folium.FeatureGroup(name='탐험가코스', overlay=True, show=True)
for index, row in df_explorer.iterrows():
    icon = CustomIcon(url("Explorer.png"), icon_size=(50, 50))
    folium.Marker(location=[row['LAT'], row['LON']], icon=icon).add_to(f_explorer)
m.add_child(f_explorer)

plugins.LocateControl().add_to(m) # 위치 컨트롤 추가
folium.LayerControl().add_to(m) # 레이어 컨트롤 추가
m
```

```python
# 타일 유형
m = folium.Map(location=[df.LAT.mean(), df.LON.mean()], zoom_start=16, tiles="CartoDB positron")

# Explorer(탐험가코스)
f_explorer = folium.FeatureGroup(name='탐험가코스', overlay=True, show=True)
for index, row in df_explorer.iterrows():
    icon = CustomIcon(url("Explorer.png"), icon_size=(50, 50))
    folium.Marker(location=[row['LAT'], row['LON']], icon=icon).add_to(f_explorer)
m.add_child(f_explorer)

# Challenger(도전자코스)
f_challenger = folium.FeatureGroup(name='도전자코스', overlay=True, show=False)
for index, row in df_challenger.iterrows():
    icon = CustomIcon(url("Challenger.png"), icon_size=(50, 50))
    folium.Marker(location=[row['LAT'], row['LON']], icon=icon).add_to(f_challenger)
m.add_child(f_challenger)

# Children(어린이코스)
f_children = folium.FeatureGroup(name='어린이코스', overlay=True, show=False)
for index, row in df_children.iterrows():
    icon = CustomIcon(url("Children.png"), icon_size=(50, 50))
    folium.Marker(location=[row['LAT'], row['LON']], icon=icon).add_to(f_children)
m.add_child(f_children)

plugins.LocateControl().add_to(m) # 위치 컨트롤 추가
folium.LayerControl().add_to(m) # 레이어 컨트롤 추가
m
```

```python
# 파일로 저장
m.save("index.html")
```
